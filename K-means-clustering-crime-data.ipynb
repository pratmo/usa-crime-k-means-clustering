# Normalization function 

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaled_crime_df = scaler.fit_transform(df.iloc[:,1:])
```

```python
scaled_crime_df
```

```python
# How to find optimum number of clusters
#The K-means algorithm aims to choose centroids that minimise the inertia, or within-cluster sum-of-squares criterion:
```

# Identify the best k-value using the elbow method

```python
from sklearn.cluster import KMeans
wcss = []
for i in range(1, 11):
    kmeans = KMeans(n_clusters=i,random_state=0)
    kmeans.fit(scaled_crime_df)
    wcss.append(kmeans.inertia_) #'within clusters sum of squares' OR intertia - this is the avg. distance of 
    # points within each cluster and store it in list wcss =[] declared above.
    
plt.plot(range(1, 11), wcss)
plt.title('Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.show()
```

# So k = 4 is the best we can see

```python
#Build Cluster algorithm
from sklearn.cluster import KMeans
clusters_new = KMeans(4, random_state=42) #randomly divide data into 4 partitions. Use any number but keep consistent
clusters_new.fit(scaled_crime_df)
clusters_new.labels_
```

```python
df['clusterid_new'] = clusters_new.labels_
```
